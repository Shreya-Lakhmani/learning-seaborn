# Box plots

 A box plot also known as box-and-whisker plot
 essential has three components:
 Box(Interqaurtile range or IQR): the box represents the interquartile , which is the range between the first quartile(Q1) and the third quartile(Q3). the length of the box signifies the spread of the middle 50% of the data.
 Line inside the box(Median): A line inside the box represents the median(Q2) which is the middle value of the dataset when it is sorted in ascending order.
 Whiskers : Whiskers extend from the box to the maximum and minimum values within a specified range(usually)
 Outliers : Individual points beyond the whiskers are considered outliers. They are potential data points that fail significantly outside the overall distribution.


```python
# Quantiles : In pandas, Quantiles can be computed using the Quantile() method, which providea a way to divide a dataset
# into intervals based on the distribution of data. Quantiles are point in dataset that correspond to certain probabilities. 
# the describe() method gives you quartiles by default.
```

```python
# Kernel Density Estimation (KDE) plots
```

```python
# The KDE plot is a smooth and continuous estimate of a histogram of a numerical variable.
# The Y-axis represents the probablity density instead of the frequency of the X-value
```

```python
plt.figure(figsize = (8, 4))
plt.subplot(1,2,1)
sns.histplot(data = df, x = "horsepower")
plt.title("Distribution of horsepower of cars", fontsize =12)
plt.xlabel("horsepower", fontsize =10)
plt.ylabel("frequency", fontsize = 10)
plt.subplot(1,2,2)
sns.kdeplot(data = df, x = "horsepower", fill = True)
plt.title("Distribution of horsepower of cars", fontsize = 12)
plt.xlabel("horsepower", fontsize = 10)
plt.ylabel("probability density", fontsize = 10)
plt.tight_layout();
```

```python
plt.figure(figsize = (8, 4))
sns.histplot(data = df, x = "horsepower", kde = True)
plt.title("Distribution of horsepower of cars", fontsize =12)
plt.xlabel("horsepower", fontsize =10)
plt.ylabel("frequency", fontsize = 10)
plt.tight_layout();
```

# Heat Maps

heat maps are visualizations that map numerical scales to color scales. This help in visualizing areas of larger numerical values and smaller numerical values. One of the most common metric that is visualized using a heat map is the Correlation coefficient.

Correlation: It is a statisticl measure that quantifies the strength and direction of a linear relationship between two variables. It is commonly expressed as the correlation coefficient which ranges from -1 to 1. values closer to 0 indicate weak relationships while values farther away from 0 indicate stronger relationships. A negative value indicates an inverse relationship while a positive value indicates a direct relationship.  

```python
df.head()
```

```python
df_num_sub = df[["curbweight", "horsepower", "citympg", "highwaympg", "price"]]
df_num_sub.head(2)
```

```python
df_num_sub =df_num_sub.convert_dtypes()
df_num_sub.info()
```

```python
corr_nat = df_num_sub.corr()
corr_nat
```

```python
corr_nat.round(2)
```

# Is there any correlation between numerical variables in the car prices dataset ? 

```python
plt.figure(figsize = (8,6))
sns.heatmap(data = corr_nat)
plt.title("Heat map of correlation of select numerical features in the car prices dataset")
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(data = corr_nat, annot = True, cmap="coolwarm", fmt=".2f", linewidths = 0.5)
plt.title("Heat map of correlation of select numerical features in the car prices dataset")
```
